```python
import matplotlib.pyplot as plt
```

```python
# Crosswalk.mp4 video, nano model
fps = [1, 2, 5, 10, 15, 25, 30]
accuracy = [0.1232, 0.2328, 0.2806, 0.306, 0.3131, 0.3247, 0.3295]
accuracy_percentage = [x / 0.3295 * 100 for x in accuracy]

plt.xlabel('FPS')
plt.ylabel('Percentage')
plt.title('Percentage of Accuracy in Static Video (compared to 30 fps)')
plt.grid(True, axis='y')
plt.plot(fps, accuracy_percentage, 'o-')

# ny_driving video, nano model
fps = [1, 2, 5, 10, 15, 25, 30]
accuracy = [0.3397, 0.3914, 0.4454, 0.4672, 0.4738, 0.4809, 0.4841]
accuracy_percentage = [x / 0.4841 * 100 for x in accuracy]

plt.xlabel('FPS')
plt.ylabel('Percentage')
plt.title('Percentage of Full Accuracy in Static Video (compared to 30 fps)')
plt.grid(True, axis='y')
plt.plot(fps, accuracy_percentage, 'o-')
```
